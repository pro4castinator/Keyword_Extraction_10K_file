```python
# First, you'll need to install the necessary libraries if you haven't already.
!pip install torch
!pip install transformers
!pip install bert-extractive-summarizer

import os
from summarizer import Summarizer

# Load the BERT extractive summarizer
model = Summarizer()

# Directory containing the reports
data_directory = '/content/data/'

# List all the report files in the directory
report_files = os.listdir(data_directory)

# Initialize an empty dictionary to store summaries
report_summaries = {}

# Loop through each report file and generate the summary
for file in report_files:
    file_path = os.path.join(data_directory, file)
    with open(file_path, 'r') as f:
        report_text = f.read()
        summary = model(report_text, min_length=60, max_length=600)  # You can adjust min_length and max_length as needed
        report_summaries[file] = summary

# Now, report_summaries contains the extractive summaries of each report
# You can further analyze the data, perform clustering, or conduct time-series analysis to answer your main research questions.
```

```python
# Print the report summaries
for file, summary in report_summaries.items():
    print(f"Summary of {file}:")
    print(summary)
    print("\n")
```

```python
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Define your custom stopwords
custom_stopwords = [
    "visa",  # Company name
    "inc",   # Common abbreviation
    "company",  # Generic term
    "financial",  # Common term
    "cik",
    "filing",
    "form",
    "tax",
    "fiscal",
]

# Add your custom stopwords to spaCy's default stopwords
nlp.Defaults.stop_words.update(custom_stopwords)

# You can also remove stopwords from spaCy's default list if needed
# Example: nlp.Defaults.stop_words.remove("word_to_remove")


# Function for text preprocessing
def preprocess_text(text):
    # Lowercase
    text = text.lower()

    # Tokenization and stopword removal
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]

    # Remove single-character tokens
    tokens = [token for token in tokens if len(token) > 1]

    # Join tokens back into text
    preprocessed_text = ' '.join(tokens)

    return preprocessed_text

# Specify the file path
file_path = '/content/data/10k2018.txt'

# Read the text from the file
with open(file_path, 'r', encoding='utf-8') as file:
    document = file.read()

# Preprocess the document
preprocessed_document = preprocess_text(document)

# Print the preprocessed document
print(preprocessed_document)
```

```python
from keybert import KeyBERT

# Read the content of the document from the file

# Initialize the KeyBERT model
kw_model = KeyBERT()

# Extract keywords with unigrams and bigrams
keywords = kw_model.extract_keywords(preprocessed_document, keyphrase_ngram_range=(1, 1), stop_words='english', top_n=20)
```

```python
print(keywords)
```

```python
!pip install nltk
import nltk
nltk.download('punkt')
# Import necessary libraries
from nltk.tokenize import sent_tokenize

# Read the content of the document from the file
file_path = '/content/data/10k2018.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    document = file.read()

# Tokenize the document into sentences
sentences = sent_tokenize(document)

# Store the sentences in the 'docs' list
docs = sentences

# Print the sentences in the 'docs' list
for doc in docs:
    print(doc)
```

```python
print(docs)
```

```python
import re

# Assuming you already have 'doc', the list of strings from the document

# Remove empty strings
docs = [line.strip() for line in docs if line.strip()]

# Define a regular expression to check for strings with only numbers, empty spaces, and special characters
pattern = re.compile(r'^[\d\s\W]+$')

# Remove strings with only numbers, empty spaces, and special characters
docs = [line for line in docs if not pattern.match(line)]

# Print the cleaned list of strings
print(docs)
```

```python
from bertopic import BERTopic

# Train a BERTopic model
topic_model = BERTopic()
topics, probs = topic_model.fit_transform(docs)
```

```python
topic_model.get_topic_info()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer_model = CountVectorizer(ngram_range=(1, 3), stop_words="english")
topic_model.update_topics(docs, vectorizer_model=vectorizer_model)
```

```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer

# Prepare embeddings
sentence_model = SentenceTransformer("distilbert-base-nli-mean-tokens")
embeddings = sentence_model.encode(docs, show_progress_bar=False)

# Create topic model
model = BERTopic()
topics, probabilities = model.fit_transform(docs, embeddings)
```

```python
from bertopic import BERTopic
from hdbscan import HDBSCAN

hdbscan_model = HDBSCAN(min_cluster_size=5, metric='euclidean', cluster_selection_method='eom', prediction_data=True)
topic_model = BERTopic(hdbscan_model=hdbscan_model)
topics, probs = topic_model.fit_transform(docs)
```
