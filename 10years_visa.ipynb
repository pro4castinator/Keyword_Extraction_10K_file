```python
import os
import pandas as pd

# Create an empty DataFrame to store the data
data_df = pd.DataFrame(columns=['data', 'year'])

# Directory where the files are located
directory = '/content/'

# Iterate through files in the directory
for filename in os.listdir(directory):
    if filename.startswith('visa_') and filename.endswith('.txt'):
        # Extract the year from the file name
        year = int(filename.split('_')[1].split('.')[0])

        # Read the text from the file
        with open(os.path.join(directory, filename), 'r', encoding='utf-8') as file:
            text = file.read()

        # Append data to the DataFrame
        data_df = data_df.append({'data': text, 'year': year}, ignore_index=True)

# Save the DataFrame to a CSV file
data_df.to_csv('/content/visa_dataset.csv', index=False)
```

```python
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Define your custom stopwords
custom_stopwords = [
    "visa",  # Company name
    "inc",   # Common abbreviation
    "company",  # Generic term
    "financial",  # Common term
    "cik",
    "filing",
    "form",
    "tax",
    "fiscal",
    "income",
    "payment",
    "volume",
]

# Add your custom stopwords to spaCy's default stopwords
nlp.Defaults.stop_words.update(custom_stopwords)

# You can also remove stopwords from spaCy's default list if needed
# Example: nlp.Defaults.stop_words.remove("word_to_remove")


# Function for text preprocessing
def preprocess_text(text):
    # Lowercase
    text = text.lower()

    # Tokenization and stopword removal
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]

    # Remove single-character tokens
    tokens = [token for token in tokens if len(token) > 1]

    # Join tokens back into text
    preprocessed_text = ' '.join(tokens)

    return preprocessed_text
```

```python
import pandas as pd

# Read the CSV file containing the dataset
data_df = pd.read_csv('/content/visa_dataset.csv')

# Combine all the text from the "data" column into a single string
combined_text = ' '.join(data_df['data'])

# Preprocess the document
preprocessed_document = preprocess_text(combined_text)

# Print the preprocessed document
print(preprocessed_document)
```

```python
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize

# Download the 'punkt' resource (only needs to be done once)
nltk.download('punkt')

# Read the CSV file containing the dataset
data_df = pd.read_csv('/content/visa_dataset.csv')

# Extract the "data" and "year" columns as lists
text_list = data_df['data'].tolist()
year_list = data_df['year'].tolist()

# Initialize lists to store sentences and their corresponding years
sentences = []
years = []

# Tokenize each document into sentences and associate with years
for i, text in enumerate(text_list):
    # Tokenize the text into sentences
    text_sentences = sent_tokenize(text)

    # Associate each sentence with the corresponding year
    year = year_list[i]
    sentences.extend(text_sentences)
    years.extend([year] * len(text_sentences))

# Create a new DataFrame with sentences and their corresponding years
new_data_df = pd.DataFrame({'Year': years, 'Sentence': sentences})

# You can now preprocess the data as needed. For example, converting to lowercase:
new_data_df['Sentence'] = new_data_df['Sentence'].str.lower()

# Print the first few rows of the new DataFrame
print(new_data_df.head())
```

```python
import pandas as pd
import nltk
import spacy
from nltk.tokenize import sent_tokenize

# Download the 'punkt' resource (only needs to be done once)
nltk.download('punkt')

# Load the spaCy model (you might need to download it if not already installed)
nlp = spacy.load('en_core_web_sm')

# Read the CSV file containing the dataset
data_df = pd.read_csv('/content/visa_dataset.csv')

# Define your custom stopwords
custom_stopwords = [
    "visa",  # Company name
    "inc",   # Common abbreviation
    "company",  # Generic term
    "financial",  # Common term
    "cik",
    "filing",
    "form",
    "tax",
    "fiscal",
    "income",
    "payment",
    "volume",
    "type", "date", "accession", "number", "global", "payment"
]

# Add your custom stopwords to spaCy's default stopwords
nlp.Defaults.stop_words.update(custom_stopwords)

# Function for text preprocessing
def p_text(text):
    # Lowercase
    text = text.lower()

    # Tokenization, stopword removal, and lemmatization
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]

    # Remove single-character tokens
    tokens = [token for token in tokens if len(token) > 1]

    # Join tokens back into text
    preprocessed_text = ' '.join(tokens)

    return preprocessed_text

# Apply text preprocessing to your DataFrame
new_data_df['Preprocessed_Sentence'] = new_data_df['Sentence'].apply(p_text)

# Print the first few rows of the preprocessed DataFrame
print(new_data_df.head())
```

```python
# Assuming 'new_data_df' is your preprocessed DataFrame
new_data_df.to_csv('/content/preprocessed_data.csv', index=False)
```

```python
from sentence_transformers import SentenceTransformer

# Pre-calculate embeddings
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = embedding_model.encode(output, show_progress_bar=True)
```

```python
from umap import UMAP

umap_model = UMAP(n_neighbors=10, n_components=5, min_dist=0.05, metric='cosine', random_state=42)
```

```python
from hdbscan import HDBSCAN

hdbscan_model = HDBSCAN(min_cluster_size=40,min_samples=20, metric='euclidean', cluster_selection_method='eom', prediction_data=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer_model = CountVectorizer(stop_words="english", min_df=2, ngram_range=(1, 2))
```

```python
from bertopic.representation import MaximalMarginalRelevance
from bertopic import BERTopic

# Create your representation model
representation_model = MaximalMarginalRelevance(diversity=0.7)
```

```python
from bertopic import BERTopic

topic_model = BERTopic(

  # Pipeline models
  embedding_model=embedding_model,
  umap_model=umap_model,
  hdbscan_model=hdbscan_model,
  vectorizer_model=vectorizer_model,
  representation_model=representation_model,

  # Hyperparameters
  top_n_words=10,
  verbose=True
)

# Train model
topics, probs = topic_model.fit_transform(output, embeddings)

# Show topics
topic_model.get_topic_info()
```
